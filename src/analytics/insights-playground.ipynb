```python
import json_numpy
import json
import codecs
import numpy as np

filename = "out.json"

with codecs.open(filename, "rb", encoding="utf-8") as source:
    node = json.loads(source.read(), object_hook=json_numpy.object_hook)
    npArray = np.array(list(node.items()))
```

```python
requests = np.array(np.array(npArray).transpose()[1], dtype=np.float64) 
print(requests)

median = np.median(requests)
stdev = np.std(requests)
print(median)
print(stdev)
```

```python
import matplotlib.pyplot as plt

plt.hist(requests, bins=100, density=True)
plt.title("Requested percentage distribution")
plt.xticks([n/10 for n in range(1,10)])

figure = plt.gcf()
```
